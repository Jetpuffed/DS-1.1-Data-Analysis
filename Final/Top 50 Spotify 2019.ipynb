```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("top50spotify.csv", encoding="ISO-8859-1")
```

```python
df.drop('Unnamed: 0', inplace=True, axis=1)
df.rename(columns={"Track.Name": "track",
                   "Artist.Name": "artist",
                   "Genre": "genre",
                   "Beats.Per.Minute": "bpm",
                   "Energy": "energy",
                   "Danceability": "danceability",
                   "Loudness..dB..": "dynamic_range",
                   "Liveness": "liveness",
                   "Valence.": "valence",
                   "Length.": "length",
                   "Acousticness..": "acousticness",
                   "Speechiness.": "speechiness",
                   "Popularity": "popularity"
                  }, inplace=True)
```

### Dataset Statistics

Count - Amount of values in the column.

Mean - Mean value of the values in the column.

Std - Standard deviation of the values in the column.

Min - Minimum value of the column.

25%/50%/75% - Percentiles of the column.

Max - Maximum value of the column.

```python
df.describe()
```

### BPM

```python
plt.figure(figsize=(10, 5))
sns.distplot(df["bpm"], kde=False, bins=16)
plt.show()

plt.figure(figsize=(10, 5))
sns.boxplot(x="genre", y="bpm", data=df)
plt.xticks(rotation=90)
plt.show()
```
